```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3, random_state=42)
data['cluster']=kmeans.fit_predict(data[['Latitude','Longitude']])
plt.scatter(data['Latitude'],data['Longitude'],c=data['cluster'])
plt.show()
```
