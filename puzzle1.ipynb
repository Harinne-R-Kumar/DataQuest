```python
import pandas as pd
emp_data=pd.read_csv("EmployeeData.csv")
print(emp_data)
dep_data=pd.read_csv("DepartmentData.csv")
print(dep_data)
```

```python
merged_df = pd.merge(emp_data, dep_data, on='department_id')
print(merged_df)
```

```python
merged_df["sal"]=merged_df['bonus'] + merged_df['base_salary']
merged_df["final"]=(merged_df['bonus'] + merged_df['base_salary'])-((merged_df['bonus'] + merged_df['base_salary'])* merged_df["tax_rate"])
print(merged_df)
```

```python
avg=pd.DataFrame(merged_df.groupby('department_name').agg(mean=('final', 'mean')))
avg.sort_values(by=['mean'], ascending=False)
```
